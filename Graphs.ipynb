```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data from the CSV file
penguins_data = pd.read_csv('penguins.csv')

# Create a scatter plot to visualize the relationship between penguin type, flipper length, and body mass
plt.figure(figsize=(12, 8))
sns.scatterplot(x='flipper_length_mm', y='body_mass_g', hue='species', data=penguins_data, palette='viridis', s=100)

# Customize the plot
plt.title('Relationship between Flipper Length, Body Mass, and Penguin Type')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Body Mass (g)')
plt.legend(title='Penguin Type')

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data from the CSV file
penguins_data = pd.read_csv('penguins.csv')

# List of unique species in the dataset
species_list = penguins_data['species'].unique()

# Create separate scatter plots for each species
for species in species_list:
    species_data = penguins_data[penguins_data['species'] == species]

    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='flipper_length_mm', y='body_mass_g', hue='sex', data=species_data, palette='viridis', s=100)
    
    # Customize the plot
    plt.title(f'{species} Penguins: Flipper Length vs Body Mass by Sex')
    plt.xlabel('Flipper Length (mm)')
    plt.ylabel('Body Mass (g)')
    plt.legend(title='Sex')
    plt.show()
```
